```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('Mall_Customers.csv')
data.head()
```

```python
plt.scatter(x=data['Annual Income (k$)'],y=data['Spending Score (1-100)'])

plt.xlabel('Income')
plt.ylabel('Spending Score')
```

```python
income = data['Annual Income (k$)']
spending_score = data['Spending Score (1-100)']

X = data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
model = KMeans(n_clusters=5)
model.fit(X)
```

```python
model.predict(X)
```

```python
model.labels_
```

```python
plt.scatter(x=data['Annual Income (k$)'],
            y=data['Spending Score (1-100)'],
            c=model.labels_,
           edgecolor='black')

plt.xlabel('Income')
plt.ylabel('Spending Score')
```

```python
model.inertia_
```

```python
k_value = []
wcss = []

for i in range(1,31):
    model = KMeans(n_clusters=i)
    model.fit(X)
    score = model.inertia_

    k_value.append(i)
    wcss.append(score)
```

```python
plt.plot(k_value,wcss)
plt.xlabel('Clusters')
plt.ylabel('WCSS')

plt.xticks(np.arange(1,31))

plt.show()
```
